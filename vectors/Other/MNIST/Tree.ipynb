```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, segmentation, color
from skimage import graph
from skimage.util import img_as_float

# Load the image
image_path = 'descarga.jpg'
image = img_as_float(io.imread(image_path))

# Generate superpixels using SLIC
num_segments = 200  # Number of superpixels
compactness = 10    # Controls the compactness of the superpixels
segments = segmentation.slic(image, n_segments=num_segments, compactness=compactness)

# Display the superpixel segmentation
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Superpixel Segmentation')
plt.imshow(segmentation.mark_boundaries(image, segments))
plt.axis('off')

plt.show()

# Create a Region Adjacency Graph (RAG)
rag = graph.rag_mean_color(image, segments)

# Display the RAG
#plt.figure(figsize=(8, 8))
#plt.title('Region Adjacency Graph')
graph.show_rag(segments, rag, image)
#plt.axis('off')
#plt.show()

# Print the edges and their weights in the RAG
print("Edges in the RAG:")
for edge in rag.edges():
    print(f"Edge: {edge}, Weight: {rag.edges[edge]['weight']}")
```
