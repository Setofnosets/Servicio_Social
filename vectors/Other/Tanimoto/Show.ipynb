```python
from rdkit import Chem
from rdkit.Chem import Draw

target = Chem.MolFromSmiles('C1=CC=C(C=C1)N')
Draw.MolToImage(target)
```

```python
import pandas as pd

df = pd.read_csv('tanimoto_similarities.csv')
df.head()

top =[]
top_legends = []
for i in range(10):
    best_graph = df.sort_values('similarity', ascending=False).head(10)['smile'].values[i]
    print(best_graph)
    mol = Chem.MolFromSmiles(best_graph)
    top.append(mol)
    best_graph = best_graph + ' Distance: ' + str(df.sort_values('similarity', ascending=False).head(10)['similarity'].values[i])
    top_legends.append(best_graph)

#Draw.MolToImage(top[0])
Draw.MolsToGridImage(mols=top, molsPerRow=5, legends=top_legends)
```
